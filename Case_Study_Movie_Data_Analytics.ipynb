# Data Cleaning

```python
#untuk membersih unique data pada file move tittle
df['movie_title'].unique()
```

```python
#untuk melihat data diplicate
df[df.duplicated()].sort_values(by = ['movie_title'])
```

```python
#cara menampilkan data yang duplikat
df[df['movie_title']=='20,000 Leagues Under the Sea']
```

```python
#cara remove duplikasi data
df2 = df.drop_duplicates().copy()
df2
```

```python
df2.info()
```

```python
#untuk mengisi data kosong ada 2 cara
#cara yang pertama
df3 = df2.copy()

#untuk menghilangkan empty values
df3 = df3[~df3['director_name'].isna()]
df3 = df3[~df3['duration'].isna()]
df3 = df3[~df3['gross'].isna()]
df3 = df3[~df3['title_year'].isna()]
df3 = df3[~df3['language'].isna()]
df3 = df3[~df3['country'].isna()]
df3 = df3[~df3['budget'].isna()]
df3 = df3[~df3['actor'].isna()]
df3 = df3[~df3['movie_facebook_likes'].isna()]
df3
```

```python
#cara2 identifikasi null column
#fungsi isna untuk mendeteksi empty kolom, 'any' setiap kolom, axis=1 untuk kolom (kalo axis=0 untuk baris)
null_rows =df2[df2.isna().any(axis=1)]
#cara menyimpan non-null rows ke df3
df3 = df2[~df2.isna().any(axis=1)].copy()
df3
```

**CASE KE - 3 INTEGER**

```python
integer_columns = ['duration', 'gross', 'title_year', 'budget', 'movie_facebook_likes']
df3[integer_columns] = df3[integer_columns].astype(int)
df3[integer_columns]
```

```python
df3[integer_columns].info()
```

```python
df3.info()
```

```python
#copy the result to df4
df4 = df3.copy()
```

same technique like we remove empty values applied here 

```python
#now, for each columns, remove empty values
df_check = df4[~(df4['movie_facebook_likes'] == 0)].copy()
```

Quite advance data cleaning: Extracting "main genre" from genres column
You can see from df4 that the movie genre in the data quite messy:some movies only assigned 1 genre, others are multiple-genre movies.

For purpose that will made clearer shortly, i am interested with the "main genre" of the move; that is, the first genre assigned to each movies. So let's find out what is it:

```python
#split columns by the separator |
df['genres'].str.split('|')
```

```python
# We just need 1 genre, so we only takes the first genre
# x[0] merupakan index pertama
df4['genres'].str.split('|').apply(lambda x: x[0]) 
```

```python
# Create new df, called df4_genre, that add main_genre columns
df4_genre = df.copy()
df4_genre['main_genre'] = df4['genres'].str.split('|').apply(lambda x: x[0])
df4_genre.head() #check the right-most column:
```

```python
# inspect unique genres extracted
df4_genre['genres'].unique()
```
